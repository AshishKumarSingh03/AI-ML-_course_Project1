# Creating the data set

```python
import numpy as np   # importing numpy linbrary
#making empty array
X=[]
Y=[]
Z=[]
for x in np.arange (-5,5,0.004):  # loop for generating x data
    X.append(x)
for y in np.arange (-5,5,0.004):
    Y.append(y)
for x in np.arange (-5,5,0.004):
    # loop for function Z
    z = 0.5 * ((x**4 + x**4) - 16 * (x**2 + x**2)+ 5 * (x + x))
    Z.append(z)
    
```

```python
X
```

```python
Y
```

```python
Z
```

```python
# Creating a dataframe for the data generated above
import pandas as pd
data=zip(X,Y,Z)
data_to_excel=pd.DataFrame(data)
```

# Applying ANN

### #Normalising the data between  0 to  1

```python
data=data.iloc[:,0:4]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(data.values)
```

```python
scaled_data.shape
```

### Dividing the data in to inpute and output

```python
X=scaled_data[:,0:2]
y=scaled_data[:,2]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=0)
```

```python
X_train.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(18,input_dim=2,activation='relu'))
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
print(model.summary())
```

```python
model.compile(optimizer='adam',loss='MSE')
```

```python
history=model.fit(X_train,y_train,epochs=100,validation_split=0.15)
```

```python
from sklearn.metrics import r2_score
```

```python
y_train_pre=model.predict(X_train)
```

## Mean squared error for training data

```python
# finding out Mean squared error for training data
mse_train=np.square(y_train_pre - y_train).mean()
```

```python
mse_train
```

```python
#
r2_train=r2_score(y_train,y_train_pre)
r2_train
```

```python
y_test_pre=model.predict(X_test)
r2_test=r2_score(y_test,y_test_pre)
r2_test
```

```python
import matplotlib.pyplot as ptl
```

```python
ptl.plot(y_train_pre,y_train)
```

```python
ptl.plot(y_test_pre,y_test)
```
